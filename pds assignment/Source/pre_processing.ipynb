```python
import pandas as pd
data=pd.read_csv("/content/train.csv")
print(data.head(),  end='\n')

''' a) Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them'''
# check for null values
print(data.isna().sum(),  end='\n')

data_2= pd.DataFrame(data)

# (b) Removed the units from Mileage, Engine, Power and New_Price and converted them into float
data_2['Mileage'] = data_2['Mileage'].str.extract('(\d+\.\d+)').astype(float)
data_2['Engine'] = data_2['Engine'].str.replace(' CC', '').astype(float)
data_2['Power'] = data_2['Power'].str.extract('(\d+\.\d+)').astype(float)
data_2['New_Price'] = data_2['New_Price'].str.extract('(\d+\.\d+)').astype(float)

print("Solution for a\n")
'''(a)
Replace the NA values of Mileage, Engine and Power with median
The median is the best way to replace outlier values because it keeps the "middle" value of the dataset the same and doesn't get affected by outliers.
Because the number of seats is a discrete quantitative variable, the mode should be used to replace any missing values.
However, because the Newprice column has many missing values, it is better to drop it altogether, as imputing any values could result in outliers.
'''

data_2['Mileage'].fillna(data_2['Mileage'].median(),inplace=True)
data_2['Engine'].fillna(data_2['Engine'].median(),inplace=True)
data_2['Power'].fillna(data_2['Power'].median(),inplace=True)
data_2['Seats'].fillna(data_2['Seats'].mode()[0],inplace=True)
data_2.drop('New_Price', axis=1, inplace=True)

output_csv_file = '/content/cleaned_data.csv'

data_2.to_csv(output_csv_file, index=False)


data_1=pd.read_csv("/content/cleaned_data.csv")
print(data_1.isna().sum(),  end='\n')

data_1['Fuel_Type'].unique()
data_1['Transmission'].unique()
data_1['Owner_Type'].unique()

print("Solution for C\n")
# (c) Convert the above categorical variables to numerical using one hot encoding
data_1['Fuel_Type'].replace({'Diesel': 0, 'Petrol': 1, 'Electric': 2}, inplace=True)
data_1['Transmission'].replace({'Manual': 0, 'Automatic': 1}, inplace=True)
data_1['Owner_Type'].replace({'First': 1, 'Second': 2, 'Third': 3, 'Fourth & Above': 4}, inplace=True)

print(data_1.head(), end = '\n')

#Save the encoded data in separate csv file
output_csv_file = '/content/encoded_data.csv'
data_1.to_csv(output_csv_file, index=False)

# (d) Add the new column Fuel efficiency = Mileage / Odometer
'''
 fuel efficiency of each car in kilometers per liter.
 This could be a useful feature for buyers who are looking for a fuel-efficient car.
'''

data_1['Fuel_efficiency'] = data_1['Mileage'] / data_1['Kilometers_Driven']

print("Solution for (d) is\n ")

#Save the file with newly added Current_Age colmun in separate file
output_csv_file = '/content/mutated_data.csv'
data_1.to_csv(output_csv_file, index=False)

data_3=pd.read_csv("/content/mutated_data.csv")
data_3.head()
```
